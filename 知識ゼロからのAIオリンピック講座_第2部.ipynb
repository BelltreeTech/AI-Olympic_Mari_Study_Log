**if = 条件で処理を分ける**

### 比較演算子

条件を書くときに使う記号です。

- `==`：等しい（左と右が同じ）
- `!=`：等しくない（左と右が違う）
- `>=`：以上（左が右と同じか大きい）
- `<=`：以下（左が右と同じか小さい）
- `>`：より大きい
- `<`：より小さい

**注意**：`=`は代入、`==`は比較です！


````python
# ==（等しい）の例
score = 100
if score == 100:
    print("満点です！")
else:
    print("もう少しです")
````

````python
# >=の例
score = 90
if score >= 80:
    print("A評価")
else:
    print("B評価")
````

````python
score = 65
if score >= 80:
    print("A評価")
else:
    print("B評価")
````

### %（剰余演算子）

`%`は割り算のあまりを求める演算子です。

例：`10 % 3` → 10を3で割ったあまり = 1

**偶数・奇数の判定や、倍数の判定に便利です。**


````python
# 偶数か奇数かを判定
num = 7
if num % 2 == 0:  # 2で割ったあまりが0なら偶数
    print(f"{num}は偶数です")
else:
    print(f"{num}は奇数です")
````

````python
# 3の倍数かどうかを判定
num = 9
if num % 3 == 0:  # 3で割ったあまりが0なら3の倍数
    print(f"{num}は3の倍数です")
else:
    print(f"{num}は3の倍数ではありません")
````

### elif（複数の条件分岐）

3つ以上の選択肢がある場合は`elif`を使います。

**書き方：**

```python
if 条件1:
    条件1がTrueの時の処理
elif 条件2:
    条件2がTrueの時の処理
else:
    どの条件にも当てはまらない時の処理
```


````python
# 成績を3段階で判定
score = 75
if score >= 80:
    print("A評価")
elif score >= 60:  # 80未満で60以上
    print("B評価")
else:  # 60未満
    print("C評価")
````

````python
# 年齢に応じた料金を表示
age = 15
if age < 6:
    print("料金: 無料")
elif age < 18:  # 6歳以上18歳未満
    print("料金: 500円")
else:  # 18歳以上
    print("料金: 1000円")
````

## 6. 関数

**関数 = 処理に名前をつけて何度でも使える**

**書き方：**

```python
def 関数名(引数):
    処理
    return 戻り値
```

**呼び出し方：**

```python
結果 = 関数名(引数)
```


````python
# さっきの条件分岐を関数にする
# 条件分岐のセクションでは、同じコードを2回書いていた
# 関数にすると、1回書けば何度でも使える！


def grade_message(score):
    if score >= 80:
        return "A評価"
    else:
        return "B評価"


# 何度でも呼べる
print(grade_message(90))
print(grade_message(65))
print(grade_message(75))
print(grade_message(95))
````

### 総復習：FizzBuzzで学んだことを確認しよう

ここまで学んだ内容（変数、リスト、条件分岐、forループ、関数）を使って、プログラミングの定番問題「FizzBuzz」に挑戦してみましょう。

**FizzBuzzとは？**

- 数字を順番に出力する
- 3の倍数なら「Fizz」
- 5の倍数なら「Buzz」
- 3と5の両方の倍数（15の倍数）なら「FizzBuzz」
- それ以外はそのまま数字を出力


**ステップ1：まず条件分岐だけで考える**

1つの数字に対してFizzBuzzの判定をしてみましょう。


````python
# 15の場合を試してみる
num = 15

# 条件分岐で判定（順番が大事！）
# 注意：15の倍数は3の倍数でも5の倍数でもあるので、先に判定する必要があります
# もし3の倍数を先に判定すると、15の時に「Fizz」と表示されてしまいます
if num % 15 == 0:  # 15の倍数（3と5両方の倍数）
    print("FizzBuzz")
elif num % 3 == 0:  # 3の倍数
    print("Fizz")
elif num % 5 == 0:  # 5の倍数
    print("Buzz")
else:  # それ以外
    print(num)
````

**ステップ2：これを関数にする**

何度も使えるように関数にしましょう。


````python
# FizzBuzzを判定する関数
def fizzbuzz(num):
    if num % 15 == 0:
        return "FizzBuzz"
    elif num % 3 == 0:
        return "Fizz"
    elif num % 5 == 0:
        return "Buzz"
    else:
        return num


# いくつか試してみる（関数を呼び出して結果を表示）
print(fizzbuzz(3))  # Fizz
print(fizzbuzz(5))  # Buzz
print(fizzbuzz(15))  # FizzBuzz
print(fizzbuzz(7))  # 7
````

**ステップ3：forループで1から30まで出力**

リストとforループを使って、1から30までのFizzBuzzを出力しましょう。


````python
# 1から30までのFizzBuzzを実行
print("=== FizzBuzz (1-30) ===")
for i in range(1, 31):  # range(1, 31)で1から30まで繰り返す
    result = fizzbuzz(i)  # 変数resultに関数の戻り値を保存
    print(f"{i}: {result}")  # f文字列で「数字: 結果」の形式で表示
````

**ステップ4：結果をリストに保存**

FizzBuzzの結果をリストに保存してみましょう。


````python
# 結果を保存するリスト
results = []

# 1から20までのFizzBuzzを実行してリストに追加
for i in range(1, 21):  # 1から20まで繰り返す
    result = fizzbuzz(i)  # 関数を呼び出して結果を取得
    results.append(result)  # リストに結果を追加

# リストの内容を確認
print("FizzBuzzの結果:", results)
print("リストの長さ:", len(results))  # len()でリストの要素数を取得
````

**このFizzBuzzで復習したこと**

✅ **変数**：`num`, `result`, `i` などを使って値を保存  
✅ **条件分岐**：`if/elif/else` で条件によって処理を分岐  
✅ **forループ**：`range()` で繰り返し処理  
✅ **関数**：`def fizzbuzz()` で処理を再利用可能に  
✅ **リスト**：`results.append()` で結果を蓄積  
✅ **f文字列**：`print(f"{i}: {result}")` で見やすく出力

これらの基本を組み合わせることで、複雑な処理も実現できます！

## 7. pandas（データ分析ツール）

**pandas = 表形式のデータ（Excelのような表）を扱うためのツール**

AI学習では大量のデータを「表」として管理することが多いです。

pandasでは**データフレーム**という、行と列からなる表形式のデータ構造を使って、Excelのようなデータを扱います。


````python
# pandasライブラリを読み込む（pdという名前で使う）
import pandas as pd

# 辞書形式でデータを作る
# 注：通常はCSVファイルから読み込みますが、今回は練習用に一時的なデータを直接作成します
data = {
    "名前": [
        "太郎",
        "花子",
        "次郎",
        "美咲",
        "健太",
        "さくら",
        "翔太",
        "咲",
        "大輔",
        "優香",
    ],
    "年齢": [25, 30, 22, 28, 24, 26, 29, 23, 27, 25],
    "得点": [85, 92, 78, 88, 76, 90, 82, 95, 79, 87],
}

# DataFrameを作る（表のこと）
df = pd.DataFrame(data)

# 表を表示
print(df)
````

### 列の演算

列に対して計算を行うこともできます。


````python
# 2つの列を使って計算（例：年齢と得点の合計）
df["年齢と得点の合計"] = df["年齢"] + df["得点"]
print("新しい列を追加:")
print(df[["名前", "年齢", "得点", "年齢と得点の合計"]].head())
````

````python
# 列の全ての値に同じ数を足す（例：得点に5点ボーナス）
df["ボーナス後の得点"] = df["得点"] + 5
print("全員に5点ボーナスを追加:")
print(df[["名前", "得点", "ボーナス後の得点"]].head())
````

### pandas のまとめ

- **DataFrame**：表形式のデータ
- **列の選択**：`df["列名"]` または `df[["列1", "列2"]]`
- **条件で絞り込み**：`df[df["列名"] >= 値]`
- **集計**：`.mean()`, `.sum()`, `.max()`, `.min()`
- **列の演算**：
  - 列同士の計算：`df["新列"] = df["列1"] + df["列2"]`
  - 列全体に同じ数を加算：`df["新列"] = df["列名"] + 数値`

## 9. AI学習デモ：Titanic生存予測

### Titanicとは？

**Titanic号**は、1912年に沈没した豪華客船です。約2200人が乗っており、そのうち約1500人が亡くなりました。
この事故のデータは、「どんな人が生き残りやすかったか」を分析するために、AI学習の教材としてよく使われています。

### やること

Titanicのデータから、生存/死亡を予測するAIを作ります。


````python
# ライブラリの読み込み
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import lightgbm as lgb
````

````python
# データを読み込む
# TitanicデータセットをURLから読み込む
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# 先頭を確認
print("データの形:", df.shape)
df.head()
````

````python
# 特徴量を選ぶ
# データを見て、予測に使えそうな情報を選びます

# 使う特徴量：
# - Pclass：客室クラス（1=上級、2=中級、3=下級）
# - SibSp：同乗している兄弟/配偶者の数
# - Parch：同乗している親/子供の数
# - Fare：運賃（チケット代）

feature_cols = [
    "Pclass",  # 客室クラス（1=上級、2=中級、3=下級）
    "SibSp",  # 同乗している兄弟/配偶者の数
    "Parch",  # 同乗している親/子供の数
    "Fare",  # 運賃（チケット代）
]
X = df[feature_cols]
y = df["Survived"]

# 学習用とテスト用に分ける
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

print("学習データ:", X_train.shape)
print("テストデータ:", X_test.shape)
````

````python
# モデルを作って学習
model = lgb.LGBMClassifier(random_state=42)
model.fit(X_train, y_train)

# 予測
pred = model.predict(X_test)

# 正解率
acc = accuracy_score(y_test, pred)
print("正解率:", acc)
````

````python
# 新しい乗客で予測
new_passenger = pd.DataFrame(
    [
        {
            "Pclass": 3,
            "SibSp": 0,
            "Parch": 0,
            "Fare": 7.25,
        }
    ]
)

prediction = model.predict(new_passenger)[0]
print("予測結果:", prediction, "（0=死亡、1=生存）")
````

### モデルの予測結果を可視化して理解を深めよう

学習したモデルの**予測結果**をグラフで確認してみましょう。
テストデータに対する予測がどれくらい当たっているのか、どんな間違いをしているのかを見てみます。


````python
# 混同行列で予測結果を分析
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

# テストデータに対する予測結果と実際の結果を比較
cm = confusion_matrix(y_test, pred)

# 混同行列を可視化
fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(cm, cmap="Blues")

# 軸ラベルとタイトル
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels(["Died (0)", "Survived (1)"])
ax.set_yticklabels(["Died (0)", "Survived (1)"])
ax.set_xlabel("Predicted", fontsize=12)
ax.set_ylabel("Actual", fontsize=12)
ax.set_title("Confusion Matrix: Model Prediction Results", fontsize=14)

# 各セルに数値を表示
for i in range(2):
    for j in range(2):
        text = ax.text(
            j, i, cm[i, j], ha="center", va="center", color="black", fontsize=16
        )

plt.colorbar(im, ax=ax)
plt.show()

# 結果の説明
print("=== 予測結果の詳細 ===")
print(f"✅ 正しく「死亡」と予測: {cm[0,0]}人")
print(f"✅ 正しく「生存」と予測: {cm[1,1]}人")
print(f"❌ 実際は死亡だが「生存」と予測: {cm[0,1]}人（偽陽性）")
print(f"❌ 実際は生存だが「死亡」と予測: {cm[1,0]}人（偽陰性）")
print()
print(f"正解率: {acc:.2%}")
````

### 実際にコードを改造してみよう

上のGPTの回答例で教えてもらった2つの改善案を実際に試してみましょう。

**やること：**

1. 元の正解率を確認（6章で実行済み）
2. 改善案①：家族の人数（FamilySize）特徴量を追加
3. 改善案②：欠損値を処理
4. 精度が向上したか比較


````python
# =========================
# 改善案①：新しい特徴量を作る
# =========================

# データを再読み込み（新しくやり直すため）
df_improved = pd.read_csv(url)

# 家族の人数を計算する新しい特徴量を追加
df_improved["FamilySize"] = df_improved["SibSp"] + df_improved["Parch"] + 1

# 確認：FamilySize列が追加されたか見てみる
print("新しい特徴量を追加しました")
print(df_improved[["SibSp", "Parch", "FamilySize"]].head())
````

````python
# =========================
# 改善案②：欠損値を処理する
# =========================

# 欠損値があるか確認
print("欠損値の数:")
print(df_improved[["Pclass", "SibSp", "Parch", "Fare", "FamilySize"]].isnull().sum())
print()

# Fareの欠損値を平均値で埋める
df_improved["Fare"] = df_improved["Fare"].fillna(df_improved["Fare"].mean())

print("欠損値を処理しました")
print("処理後の欠損値の数:")
print(df_improved[["Pclass", "SibSp", "Parch", "Fare", "FamilySize"]].isnull().sum())
````

````python
# =========================
# 改善したモデルで学習・評価
# =========================

# 改善版の特徴量（FamilySizeを追加）
feature_cols_improved = [
    "Pclass",
    "SibSp",
    "Parch",
    "Fare",
    "FamilySize",  # 追加した特徴量
]

X_improved = df_improved[feature_cols_improved]
y_improved = df_improved["Survived"]

# 学習用とテスト用に分ける
X_train_improved, X_test_improved, y_train_improved, y_test_improved = train_test_split(
    X_improved, y_improved, test_size=0.25, random_state=42
)

# モデルを作って学習
model_improved = lgb.LGBMClassifier(random_state=42, verbose=-1)
model_improved.fit(X_train_improved, y_train_improved)

# 予測
pred_improved = model_improved.predict(X_test_improved)

# 正解率
acc_improved = accuracy_score(y_test_improved, pred_improved)

# 比較
print("【精度の比較】")
print(f"元の正解率:     {acc:.4f} (特徴量: Pclass, SibSp, Parch, Fare)")
print(
    f"改善後の正解率: {acc_improved:.4f} (特徴量: Pclass, SibSp, Parch, Fare, FamilySize + 欠損値処理)"
)
print(f"改善度:         {acc_improved - acc:+.4f}")
print()
if acc_improved > acc:
    print("✅ 精度が向上しました！")
else:
    print("❌ 今回は精度が下がってしまいましたが、データや設定によって結果は変わります")
````

> - どれくらい慎重に学習するか
>
> を **自分で調整しないと本領を発揮しません**。
>
> ---
>
> ### 初学者向けの改善例
>
> まずは「考え方が分かりやすいパラメータ」だけ触ります。
>
> ```python
> model = lgb.LGBMClassifier(
>     random_state=42,
>     n_estimators=30,   # 木の本数（多いほど学習量が増える）
>     learning_rate=0.05  # 学習の慎重さ（小さいほど丁寧）
> )
> ```
>
> ---
>
> ### イメージで説明すると
>
> - `n_estimators`
>   → **何回勉強するか**
> - `learning_rate`
>   → **1回の勉強をどれくらい丁寧にやるか**
>
> 👉 少しずつ丁寧に学習させると、
> 雑な判断が減って精度が上がりやすくなります。
>
> ---
>
> ## 改善案②：評価方法を変える（1回の分割に頼らない）
>
> ### 今の状態
>
> ```python
> train_test_split(..., random_state=42)
> ```
>
> これは **1回だけ** データを分けて評価しています。
>
> ---
>
> ### なぜこれが問題？
>
> たまたま
>
> - 学習しやすい分け方
> - 学習しにくい分け方
>
> になることがあります。
>
> 👉 **運に左右されやすい**のです。
>
> ---
>
> ### 改善方法：交差検証（超重要）
>
> 「分け方を変えて何回も評価する」方法です。
>
> ```python
> from sklearn.model_selection import cross_val_score
>
> model = lgb.LGBMClassifier(random_state=42)
>
> scores = cross_val_score(
>     model, X, y,
>     cv=5,            # 5回に分けて評価
>     scoring="accuracy"
> )
>
> print("各回の正解率:", scores)
> print("平均正解率:", scores.mean())
> ```
>
> ---
>
> ### 初学者向けイメージ
>
> - テストを **1回だけ受ける** → 運の要素あり
> - テストを **5回受けて平均を見る** → 実力が分かる
>
> 👉 モデルの良し悪しを **正しく判断** できるようになります。
>
> ---
>
> ## まとめ（今回の結論）
>
> | 改善案  | 何を変える？           | 効果                 |
> | ------- | ---------------------- | -------------------- |
> | 改善案① | モデルのパラメータ調整 | 学習の質が上がる     |
> | 改善案② | 評価方法を改善         | 精度の判断が安定する |
>
> ---
>
> ### とても大事なこと
>
> 今回のように
>
> > 特徴量を足しても改善しない
>
> これは
> ❌ 失敗
> ではなく
> ✅ **「もうモデル側を調整する段階に来た」サイン**
> です。
>
> この流れは **実務のAI改善とまったく同じ**です 👏

</details>


### パラメータ調整を試してみよう

GPTの回答にあった「改善案①：パラメータ調整」を実際に試してみます。


````python
# =========================
# パラメータを調整したモデル
# =========================

# 元のデータを使う（特徴量追加なし）
X_tuned = df[feature_cols]
y_tuned = df["Survived"]

# 学習用とテスト用に分ける
X_train_tuned, X_test_tuned, y_train_tuned, y_test_tuned = train_test_split(
    X_tuned, y_tuned, test_size=0.25, random_state=42
)

# パラメータを調整したモデル
model_tuned = lgb.LGBMClassifier(
    random_state=42,
    n_estimators=30,  # 木の本数（多いほど学習量が増える）
    learning_rate=0.05,  # 学習の慎重さ（小さいほど丁寧）
    verbose=-1,  # 学習ログを非表示
)

model_tuned.fit(X_train_tuned, y_train_tuned)

# 予測
pred_tuned = model_tuned.predict(X_test_tuned)

# 正解率
acc_tuned = accuracy_score(y_test_tuned, pred_tuned)

# 比較
print("【パラメータ調整の効果】")
print(f"元の正解率（デフォルト設定）:      {acc:.4f}")
print(f"パラメータ調整後の正解率:          {acc_tuned:.4f}")
print(f"改善度:                           {acc_tuned - acc:+.4f}")
print()
if acc_tuned > acc:
    print("✅ パラメータ調整で精度が向上しました！")
else:
    print(
        "今回は大きな改善は見られませんでしたが、データセットによって効果は異なります。"
    )
````
